**PROBABLIITY**

```python
import numpy as np

# Generate sample heights data (you can replace this with your dataset)
np.random.seed(0)  # for reproducibility
heights = np.random.normal(loc=170, scale=10, size=1000)  # mean=170, std=10, 1000 samples

# Define a function to calculate the probability of a person being taller than a certain height
def probability_taller_than(height_data, threshold_height):
    taller_count = sum(height > threshold_height for height in height_data)
    total_count = len(height_data)
    probability = taller_count / total_count
    return probability

# Define a threshold height
threshold = 180  # Change this to your desired threshold height

# Calculate the probability of a person being taller than the threshold height
prob_taller_than_threshold = probability_taller_than(heights, threshold)

# Print the result
print(f"The probability of a person being taller than {threshold} cm is: {prob_taller_than_threshold:.2f}")
```

```python
import numpy as np
from scipy.stats import t

# Generate sample data (heights and weights)
np.random.seed(0)  # For reproducibility
heights = np.random.normal(loc=170, scale=10, size=1000)  # Mean=170, Std=10
weights = np.random.normal(loc=70, scale=5, size=1000)   # Mean=70, Std=5

# Function to calculate confidence interval
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    margin_error = stderr * t.ppf((1 + confidence) / 2, n - 1)
    return mean - margin_error, mean + margin_error

# Calculate confidence interval for heights
heights_ci = confidence_interval(heights)
print("95% Confidence Interval for Heights Mean:", heights_ci)

# Calculate confidence interval for weights
weights_ci = confidence_interval(weights)
print("95% Confidence Interval for Weights Mean:", weights_ci)
```

#**HYPOTHESIS TESTING**

```python
import numpy as np
from scipy.stats import ttest_1samp

# Generate sample data (heights)
np.random.seed(0)  # For reproducibility
heights = np.random.normal(loc=170, scale=10, size=1000)  # Mean=170, Std=10

# Hypothesized mean height
hypothesized_mean = 175

# Perform t-test
t_statistic, p_value = ttest_1samp(heights, hypothesized_mean)

# Define significance level
alpha = 0.05

# Print t-test results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Check if null hypothesis should be rejected
if p_value < alpha:
    print("Reject null hypothesis: Mean height is significantly different from", hypothesized_mean, "cm.")
else:
    print("Fail to reject null hypothesis: Mean height is not significantly different from", hypothesized_mean, "cm.")
```

```python
import numpy as np
from scipy.stats import t
from sklearn.linear_model import LinearRegression

# Generate sample data (heights and weights)
np.random.seed(0)  # For reproducibility
heights = np.random.normal(loc=170, scale=10, size=100)  # Mean=170, Std=10
weights = np.random.normal(loc=70, scale=5, size=100)   # Mean=70, Std=5

# Reshape heights for sklearn
heights = heights.reshape(-1, 1)

# Fit linear regression model
model = LinearRegression()
model.fit(heights, weights)

# Perform hypothesis test
coeff = model.coef_[0]  # Coefficient for heights
n = len(weights)  # Sample size
df = n - 2  # Degrees of freedom
alpha = 0.05  # Significance level

# Compute standard error of coefficient estimate
stderr = np.sqrt(np.sum((weights - model.predict(heights))**2) / (n - 2)) / np.sqrt(np.sum((heights - np.mean(heights))**2))

# Compute t-statistic
t_statistic = coeff / stderr

# Compute critical value from t-distribution
critical_value = t.ppf(1 - alpha/2, df)

# Compute p-value
p_value = 2 * (1 - t.cdf(abs(t_statistic), df))

# Print results
print("Coefficient:", coeff)
print("Standard Error of Coefficient Estimate:", stderr)
print("T-statistic:", t_statistic)
print("Critical Value from t-distribution:", critical_value)
print("P-value:", p_value)

# Check if coefficient is statistically significant
if abs(t_statistic) > critical_value:
    print("Reject null hypothesis: The coefficient for heights is statistically significant.")
else:
    print("Fail to reject null hypothesis: The coefficient for heights is not statistically significant.")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import t

# Generate sample data (heights and weights)
np.random.seed(0)  # For reproducibility
heights = np.random.normal(loc=170, scale=10, size=100)  # Mean=170, Std=10
weights = np.random.normal(loc=70, scale=5, size=100)   # Mean=70, Std=5

# Reshape heights for sklearn
heights = heights.reshape(-1, 1)

# Fit linear regression model
model = LinearRegression()
model.fit(heights, weights)

# Perform hypothesis test
coeff = model.coef_[0]  # Coefficient for heights
n = len(weights)  # Sample size
df = n - 2  # Degrees of freedom
alpha = 0.05  # Significance level

# Compute standard error of coefficient estimate
stderr = np.sqrt(np.sum((weights - model.predict(heights))**2) / (n - 2)) / np.sqrt(np.sum((heights - np.mean(heights))**2))

# Compute t-statistic
t_statistic = coeff / stderr

# Compute p-value
p_value = 2 * (1 - t.cdf(abs(t_statistic), df))

# Print results
print("P-value:", p_value)

# Decision: Include or exclude variable based on p-value
if p_value < alpha:
    print("Include variable 'heights' in the model (statistically significant).")
else:
    print("Exclude variable 'heights' from the model (not statistically significant).")
```
