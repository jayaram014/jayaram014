#**Perform following operations on images**

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2

# Load the image
image = cv2.imread('test.png')

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#**Resizing to 224, 224, 3**

```python
# Resizing to 224x224x3
resized_image = cv2.resize(image, (224, 224))

cv2_imshow(resized_image)
```

#**Grayscale Conversion**

```python
# Grayscale Conversion
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray_image)
```

#**Cropping with suitable coordinates to extract only dog**

```python
# Cropping with Suitable Coordinates
x, y, w, h = 100, 100, 300, 200  # Example coordinates of the bounding box
cropped_image = image[y:y+h, x:x+w]

cv2_imshow(cropped_image)
```

#**Rotation by 45 degrees**

```python
# Rotation by 45 Degrees
rows, cols = image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (cols, rows))

cv2_imshow(rotated_image)
```

#**Flip left to right**

```python
# Flip Left to Right
flipped_image = cv2.flip(image, 1)

cv2_imshow(flipped_image)
```

#**Gaussian & Median Blurring**

```python
# Gaussian & Median Blurring
gaussian_blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
median_blurred_image = cv2.medianBlur(image, 5)

cv2_imshow(gaussian_blurred_image)
cv2_imshow(median_blurred_image)
```

#**Edge Detection techniques**

```python
# Edge Detection Techniques
edges_sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
edges_sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)
edges_sobel = cv2.magnitude(edges_sobel_x, edges_sobel_y)

edges_canny = cv2.Canny(gray_image, 100, 200)

cv2_imshow(edges_sobel.astype('uint8'))
cv2_imshow(edges_canny)
```

#**Background Subtraction**

```python
# Background Subtraction
background_subtractor = cv2.createBackgroundSubtractorMOG2()
foreground_mask = background_subtractor.apply(image)

cv2_imshow(foreground_mask)
```
